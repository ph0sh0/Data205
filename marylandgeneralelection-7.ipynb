```python
# prompt: filter out anything in office name except president -  vice president

import pandas as pd

# Read the CSV files into DataFrames, specifying the encoding as 'latin1'
combined_df = pd.read_csv('/content/drive/MyDrive/All_By_Precinct_2020_General.csv')

# Filter the DataFrame to include only rows where the "Office Name" column contains "President" or "Vice President"
filtered_df = combined_df[combined_df['Office Name'].str.contains('President|Vice President', case=False, na=False)]

# Re-displaying the filtered data for clarity
filtered_df.head()
```

```python
# prompt: sum the number of total votes for each candidate

# Group the filtered DataFrame by candidate name and sum the total votes for each candidate
candidate_votes = filtered_df.groupby('Candidate Name')['Total Votes'].sum()

# Display the total votes for each candidate
candidate_votes
```

```python
# Grouping the data by County to get the sum of total votes in each county
county_votes_df = filtered_df.groupby("County")["Total Votes"].sum().reset_index()

# Renaming columns for better readability
county_votes_df.columns = ["County", "Total Votes"]

# Sorting the data by total votes for better visualization
county_votes_df = county_votes_df.sort_values(by="Total Votes", ascending=False)

# Display the data to the user using the standard display method for DataFrames
# Instead of ace_tools, we're using the DataFrame's display functionality
print("Total Votes by County:")
display(county_votes_df) # or print(county_votes_df) if you prefer simpler output
# or
# county_votes_df.head() # to display first 5 entries
# county_votes_df.to_csv('county_votes.csv', index=False) # to store it as CSV file
```

```python
import matplotlib.pyplot as plt
# Filtering the data for the presidential election only
presidential_df = combined_df[combined_df['Office Name'].str.contains('President', case=False, na=False)]

# Aggregating votes by party for different voting types within presidential election
presidential_party_votes = presidential_df.groupby('Party').agg({
    'Early Voting Votes': 'sum',
    'Election Day Votes': 'sum',
    'By Mail Votes': 'sum',
    'By Mail 2 Votes': 'sum',
    'Prov. Votes': 'sum'
}).reset_index()

# Plotting scatter plots for each voting type within presidential election
plt.figure(figsize=(10, 6))

plt.scatter(presidential_party_votes['Party'], presidential_party_votes['Early Voting Votes'], label='Early Voting', alpha=0.6)
plt.scatter(presidential_party_votes['Party'], presidential_party_votes['Election Day Votes'], label='Election Day', alpha=0.6)
plt.scatter(presidential_party_votes['Party'], presidential_party_votes['By Mail Votes'], label='By Mail', alpha=0.6)
plt.scatter(presidential_party_votes['Party'], presidential_party_votes['By Mail 2 Votes'], label='By Mail 2', alpha=0.6)
plt.scatter(presidential_party_votes['Party'], presidential_party_votes['Prov. Votes'], label='Prov. Votes', alpha=0.6)

plt.xlabel('Party')
plt.ylabel('Total Votes')
plt.title('Scatter Plot of Total Votes by Party (Presidential Election: Mail-in, Early, and Election Day Voting)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Comparing total mail-in vs election-day turnout across all parties in the presidential election

# Summing up total mail-in and election-day votes for the presidential election
mail_in_vs_election_day = presidential_df[['By Mail Votes', 'Election Day Votes']].sum()

# Plotting the comparison as a bar chart
plt.figure(figsize=(8, 5))
mail_in_vs_election_day.plot(kind='bar', color=['blue', 'green'], alpha=0.7)
plt.ylabel('Total Votes')
plt.title('Total Mail-in vs. Election-Day Turnout (Presidential Election)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Creating a mapping dictionary for county codes to county names (based on Maryland counties, if applicable)
county_code_to_name = {
    1: "Allegany", 2: "Anne Arundel", 3: "Baltimore City", 4: "Baltimore County",
    5: "Calvert", 6: "Caroline", 7: "Carroll", 8: "Cecil", 9: "Charles",
    10: "Dorchester", 11: "Frederick", 12: "Garrett", 13: "Harford", 14: "Howard",
    15: "Kent", 16: "Montgomery", 17: "Prince George's", 18: "Queen Anne's",
    19: "St. Mary's", 20: "Somerset", 21: "Talbot", 22: "Washington",
    23: "Wicomico", 24: "Worcester"
}

# Replacing numeric county codes with actual county names
presidential_df['County'] = presidential_df['County'].map(county_code_to_name)

# Grouping data by updated county names for vote distribution
county_vote_distribution_named = presidential_df.groupby('County').agg({
    'Total Votes': 'sum'
}).reset_index()

# Plotting the distribution of votes across counties with names
plt.figure(figsize=(12, 6))
plt.bar(county_vote_distribution_named['County'], county_vote_distribution_named['Total Votes'], color='teal', alpha=0.7)
plt.xlabel('County')
plt.ylabel('Total Votes')
plt.title('Vote Distribution Across Counties (Presidential Election)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()

```

```python
# Grouping total votes by county and voting type for the presidential election
county_voting_type_totals = presidential_df.groupby('County').agg({
    'Early Voting Votes': 'sum',
    'Election Day Votes': 'sum',
    'By Mail Votes': 'sum',
    'By Mail 2 Votes': 'sum',
    'Prov. Votes': 'sum'
}).reset_index()

# Plotting turnout differences by voting type for each county
county_voting_type_totals.set_index('County').plot(kind='bar', stacked=True, figsize=(14, 7), alpha=0.7)
plt.xlabel('County')
plt.ylabel('Total Votes')
plt.title('Turnout Differences by Voting Type for Each County (Presidential Election)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Voting Type')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# URL to the GeoJSON file
geojson_url = "https://raw.githubusercontent.com/frankrowe/maryland-geojson/refs/heads/master/maryland-counties.geojson"

# Maryland counties and election results
county_results = pd.DataFrame({
    "County": [
        "Allegany", "Anne Arundel", "Baltimore", "Baltimore City", "Calvert",
        "Caroline", "Carroll", "Cecil", "Charles", "Dorchester", "Frederick",
        "Garrett", "Harford", "Howard", "Kent", "Montgomery", "Prince George's",
        "Queen Anne's", "Somerset", "St. Mary's", "Talbot", "Washington",
        "Wicomico", "Worcester"
    ],
    "Winner": [
        "REP", "DEM", "DEM", "DEM", "REP", "REP", "REP", "REP", "DEM", "REP",
        "DEM", "REP", "REP", "DEM", "DEM", "DEM", "DEM", "REP", "REP", "REP",
        "DEM", "REP", "REP", "REP"
    ]
})

# Create a choropleth map
fig = px.choropleth(
    county_results,
    geojson=geojson_url,
    locations="County",
    featureidkey="properties.name",
    color="Winner",
    color_discrete_map={"DEM": "blue", "REP": "red"},
    title="2020 Presidential Election Results by County in Maryland",
    labels={"Winner": "Winning Party"}
)

# Enhance the map visuals
fig.update_geos(fitbounds="locations", visible=False)
fig.show()




```

```python
# Aggregate total votes for Democrats and Republicans using proper column names
aggregated_votes = combined_df[
    (combined_df["Office Name"] == "President - Vice Pres")
    & (combined_df["Party"].isin(["DEM", "REP"]))
].groupby("Party")["Total Votes"].sum().reset_index()

# Rename columns for clarity
aggregated_votes.columns = ["Party", "Total Votes"]

# Create a bar chart to compare total votes for Democrats and Republicans
plt.figure(figsize=(8, 6))
plt.bar(aggregated_votes["Party"], aggregated_votes["Total Votes"], color=["blue", "red"])
plt.title("Total Votes for Democrats vs Republicans in Maryland (2020)", fontsize=16)
plt.ylabel("Total Votes", fontsize=14)
plt.xlabel("Party", fontsize=14)

# Annotate bars with vote totals
for i, v in enumerate(aggregated_votes["Total Votes"]):
    plt.text(i, v + 5000, f"{int(v):,}", ha="center", fontsize=12)

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()


```

```python
# Filter for the Presidential race and exclude 'BOT'
presidential_race = combined_df[
    (combined_df['Office Name'] == 'President - Vice Pres') &
    (combined_df['Party'] != 'BOT')
]

# Group by County and Party, summing Total Votes
grouped_presidential = presidential_race.groupby(['County', 'Party'])['Total Votes'].sum().reset_index()

# Determine the winning party for each county
winning_party_presidential = grouped_presidential.loc[
    grouped_presidential.groupby('County')['Total Votes'].idxmax()
]

# Apply the county mapping
winning_party_presidential['County'] = winning_party_presidential['County'].map(county_code_to_name)

# Filter for Democrat and Republican parties only
filtered_winning_party = winning_party_presidential[
    winning_party_presidential['Party'].isin(['DEM', 'REP'])
]

# Map colors for Democrat (blue) and Republican (red)
color_map = {'DEM': 'blue', 'REP': 'red'}
filtered_winning_party['Color'] = filtered_winning_party['Party'].map(color_map)

# Create the horizontal bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(
    filtered_winning_party['County'],
    filtered_winning_party['Total Votes'],
    color=filtered_winning_party['Color']
)

# Add labels for the parties on the bars
for bar, party in zip(bars, filtered_winning_party['Party']):
    plt.text(bar.get_width() - 5000, bar.get_y() + bar.get_height()/2, party,
             ha='right', va='center', fontsize=9, color='white')

# Customize the plot
plt.xlabel('Total Votes')
plt.ylabel('Counties')
plt.title('Winning Party (Presidential Race) by Total Votes for Each County (Horizontal)')
plt.tight_layout()
plt.show()
```
